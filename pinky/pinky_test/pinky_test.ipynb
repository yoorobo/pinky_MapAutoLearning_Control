### 초음파

```python
from pinkylib import Ultrasonic
import time

pinky = Ultrasonic()

for i in range(5):
    dist = pinky.get_dist()
    print(f"{dist:.2f} cm")
    time.sleep(0.5)
    
pinky.close()
```

### IR

```python
from pinkylib import IR
import time

pinky = IR()

for i in range(10):
    ir_1, ir_2, ir_3 = pinky.read_ir()
    print(f"ir_1: {ir_1}, ir_2: {ir_2}, ir_3: {ir_3}")
    time.sleep(0.5)
    
pinky.close()
```

### IMU

```python
from pinkylib import IMU
import time

#imu = IMU()

# 9축 모드 설정
imu = IMU(mode='9dof') 

try:
    for i in range(100):
        imu_data = imu.read_imu_data()

        acc = imu_data['acc']
        mag = imu_data['mag']
        gyro = imu_data['gyro']
        euler = imu_data['euler']

        print("*"*10)

        print(f"ACC   | X:{acc[0]:6.2f} Y:{acc[1]:6.2f} Z:{acc[2]:6.2f} (m/s^2)")
        print(f"MAG   | X:{mag[0]:6.2f} Y:{mag[1]:6.2f} Z:{mag[2]:6.2f} (uT)")
        print(f"GYRO  | X:{gyro[0]:6.2f} Y:{gyro[1]:6.2f} Z:{gyro[2]:6.2f} (deg/s)")
        print(f"EULER | Roll:{euler[0]:6.2f} Pitch:{euler[1]:6.2f} Yaw:{euler[2]:6.2f} (deg)")

        time.sleep(0.5)
   
    imu.close()
except KeyboardInterrupt:
    print("Process stopped by user")
```

### LED

```python
import time
from pinkylib import LED

RED = (255, 0, 0)
ORANGE = (255, 127, 0)
YELLOW = (255, 255, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)
INDIGO  = (75, 0, 130)
VIOLET = (148, 0, 211)
WHITE = (255, 255, 255)

leds = LED()
# 1. 모든 LED를 빨간색으로 켜기
print("모든 LED를 3초간 빨간색으로 켭니다.")
leds.fill(RED)
time.sleep(3)

# 2. 모든 LED를 초록색으로 켜기
print("\n모든 LED를 3초간 초록색으로 켭니다.")
leds.fill(GREEN)
time.sleep(3)

# 3. 모든 LED를 파란색으로 켜기
print("\n모든 LED를 3초간 파란색으로 켭니다.")
leds.fill(BLUE) 
time.sleep(3)

# 4. 무지개 출력
print("\n모든 LED를 3초간 무지개로 켭니다.")

leds.set_pixel(0, RED)
leds.set_pixel(1, ORANGE)
leds.set_pixel(2, YELLOW)
leds.set_pixel(3, GREEN)
leds.set_pixel(4, BLUE)
leds.set_pixel(5, INDIGO)
leds.set_pixel(6, VIOLET)
leds.set_pixel(7, WHITE)

leds.show()
time.sleep(3)

leds.close()
```
