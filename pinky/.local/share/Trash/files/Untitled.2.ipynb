```python
import RPi.GPIO as GPIO
import time

BUZZER_PIN = 4

C5 = 523
E5 = 659
G5 = 784

GPIO.setwarnings(False)
GPIO.setmode(GPIO.BCM)
GPIO.setup(BUZZER_PIN, GPIO.OUT)

pwm = GPIO.PWM(BUZZER_PIN, C5)    
pwm.start(90)

pwm.ChangeFrequency(C5)
time.sleep(0.08)

pwm.ChangeFrequency(E5)
time.sleep(0.08)

pwm.ChangeFrequency(G5)
time.sleep(0.3)

pwm.stop()
```
